```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/drive/MyDrive/eda/team_stats_2003_2023.csv")
df.head(5)
```

```python
# Categorical data
# countplot
sns.countplot(data=df, x='wins')
df['losses'].value_counts()
```

```python
 # pie chart

df['g'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['wins'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['penalties'])
```

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['wins'])
```
